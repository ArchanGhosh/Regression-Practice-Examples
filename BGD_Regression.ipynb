```
import numpy as np
import tensorflow as tf
```

```
class BGDRegression(object):
  def __init__(self, eta=0.1, n_iter=1000):
    self.eta=eta
    self.n_iter=n_iter
  def fit(self, X, y):
    m,n=X.shape
    ones=tf.ones(shape=[m,1], dtype=tf.dtypes.float64)
    X= tf.concat([ones, X], axis=1)

    self.theta = tf.Variable(tf.random.normal(shape=[n+1, 1], dtype=tf.dtypes.float64))
    for i in range(self.n_iter):
      gradients = 2 / m * tf.transpose(X) @ (X @ self.theta - y)
      self.theta.assign(self.theta - self.eta * gradients)
    return self
  def predict(self,X):
    m,n=X.shape
    ones=tf.ones(shape=[m,1], dtype=tf.dtypes.float64)
    X = tf.concat([ones, X], axis=1)
    return X @ self.theta 
```

```
from tensorflow.keras.datasets import boston_housing
```

```
mean = tf.reduce_mean(X_train, axis= 1, keepdims=True)
```

```
std = tf.math.reduce_std(X_train, axis=1 , keepdims=True)
```

```
X_scaled = (X_train - mean) / std
```

```
bgd_reg = BGDRegression()
```

```
X_train = tf.Variable(X_scaled) 
```

```
y_train = tf.Variable(y_train)
```

```
bgd_reg.fit(X_train, y_train)
```
